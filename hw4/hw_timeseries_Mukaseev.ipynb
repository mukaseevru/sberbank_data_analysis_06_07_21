```python
import numpy as np
import pandas as pd
import datetime as dt
```

```python
week_days_dct = {'пн': 0,
                 'вт': 1,
                 'ср': 2,
                 'чт': 3,
                 'пт': 4,
                 'сб': 5,
                 'вс': 6}
week_days_dct
```

```python
months_dct = {'январь': 1,
              'февраль': 2,
              'март': 3,
              'апрель': 4,
              'май': 5,
              'июнь': 6,
              'июль': 7,
              'август': 8,
              'сентябрь': 9,
              'октябрь': 10,
              'ноябрь': 11,
              'декабрь': 12}
months_dct
```

```python
holidays_lst = ['01.01.1999',
                '02.01.1999',
                '03.01.1999',
                '04.01.1999',
                '05.01.1999',
                '06.01.1999',
                '07.01.1999',
                '23.02.1999',
                '08.03.1999',
                '01.05.1999',
                '09.05.1999',
                '12.06.1999',
                '04.11.1999']
holidays_lst = pd.to_datetime(
    holidays_lst, format='%d.%m.%Y').strftime('%B %d')
holidays_lst
```

```python
daysoff = []
day_before_holidays = []
day_before_new_year = []
```

```python
for i in df.index:
    for month in months_dct:
        days = df.loc[i, month.capitalize()].split(',')
        for day in days:
            if '*' in day:
                day_before_holidays.append(str(day[:-1]) + ', '
                                           + str(months_dct[month]) + ', '
                                           + str(df['Год/Месяц'][i]))
            elif '+' in day:
                daysoff.append(str(day[:-1]) + ', '
                               + str(months_dct[month]) + ', '
                               + str(df['Год/Месяц'][i]))
            else:
                daysoff.append(str(day) + ', '
                               + str(months_dct[month]) + ', '
                               + str(df['Год/Месяц'][i]))
    day_before_new_year.append(daysoff[-1])
```

```python
daysoff = pd.to_datetime(daysoff, format='%d, %m, %Y')
daysoff
```

```python
day_before_holidays = pd.to_datetime(day_before_holidays, format='%d, %m, %Y')
day_before_holidays
```

```python
day_before_new_year = pd.to_datetime(day_before_new_year, format='%d, %m, %Y')
day_before_new_year
```

```python
from pandas.tseries.holiday import AbstractHolidayCalendar, Holiday, nearest_workday, EasterMonday
from pandas.tseries.offsets import DateOffset, Day


class RussianBusinessCalendar(AbstractHolidayCalendar):
    start_date = dt.datetime(1999, 1, 1)
    end_date = dt.datetime(2025, 12, 31)
    rules = [Holiday('Russian Days Off', month=day.month, day=day.day,
                     year=day.year) for day in daysoff]


russian_business_calendar = pd.offsets.CustomBusinessDay(weekmask='Mon Tue Wed Thu Fri',
                                                         calendar=RussianBusinessCalendar())
```

```python
full_events = pd.DataFrame(index=pd.date_range(
    '1999-01-01', '2025-12-31', freq='D'))
full_events['выходной'] = [int(x)
                               for x in [i in daysoff for i in full_events.index]]
full_events['праздник'] = [int(x) for x in [i.strftime(
    '%B %d') in holidays_lst for i in full_events.index]]
day_before_holidays = [dt.datetime(i.year, i.month, i.day) + pd.Timedelta(
    days=1) - russian_business_calendar for i in full_events[full_events['праздник'] == 1].index]
full_events['предпраздничный рабочий день'] = [
    int(x) for x in [i in day_before_holidays for i in full_events.index]]
last_day_of_month = [pd.to_datetime(str(year)) + pd.offsets.MonthEnd(i)
                     for year in full_events.index.year.unique() for i in range(1, 13)]
full_events['последний день месяца'] = [
    int(x) for x in [i in last_day_of_month for i in full_events.index]]
last_work_day_of_month = [dt.datetime(i.year, i.month, i.day) + pd.Timedelta(
    days=1) - russian_business_calendar for i in last_day_of_month]
full_events['последний рабочий день месяца'] = [
    int(x) for x in [i in last_work_day_of_month for i in full_events.index]]
for week_day in week_days_dct:
    full_events[week_day] = [1 if i == week_days_dct[week_day]
                             else 0 for i in full_events.index.weekday]
for month in months_dct:
    full_events[month] = [1 if i == months_dct[month]
                          else 0 for i in full_events.index.month]
full_events['предновогодний выходной день'] = [int(x) for x in [i in day_before_new_year for i in full_events.index]]
full_events
```
